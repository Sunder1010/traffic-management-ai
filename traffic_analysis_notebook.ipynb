```python
import pandas as pd

data = pd.read_csv("region_traffic.csv")

data.head()
```

```python
import matplotlib.pyplot as plt

traffic_by_region = data.groupby("region_name")["all_motor_vehicles"].sum()

traffic_by_region.sort_values(ascending=False).head(10).plot(kind="bar")

plt.title("Top 10 Regions by Traffic Volume")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

features = data[[
    "pedal_cycles",
    "two_wheeled_motor_vehicles",
    "cars_and_taxis",
    "buses_and_coaches",
    "lgvs",
    "all_hgvs"
]]

target = data["all_motor_vehicles"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

print("Model accuracy:", model.score(X_test, y_test))
```
